**Besides, there are some specific types of wordforms:**

| Wordforms | Description | Example |
|---|---|---|
| Fragments | incomplete, repitition, prolonging words before a complete word in the sentence | Miss- Mrs., main- mainly, I-I-I'd, st-st-store, etc. |
| Filled Pauses | words to fill-in between words to pause or sound natural | uhh, uhm, ahhh, hmm, you know, like, etc. |
| Lemma | words having same stem, same POS, roughly same meaning | cat→cats; they→their; We→our; He→him; better,best→good; am,is,are,was,were→be |
